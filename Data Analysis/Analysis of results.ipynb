```python
from matplotlib import pyplot as plt
import math
from sklearn.metrics import mean_squared_error
import numpy as np

content=["CG-Animation","BirdsInCage","CrowdRun","Show","Water","Animation","OldTown","MultiPeople","TennisPlayer"]

def drawScatter(sample_data,a,b=None,c=None):
    plt.figure()
    plt.xlabel('DMOS')
    plt.ylabel('VMAF_score')
    dataframe=sample_data.loc[sample_data["content_id"] == a,["DMOS","VMAF_score","VMAF_feature_motion2_score"]] 
    print(dataframe[1:].describe())
    x=list(dataframe["DMOS"])
    y=list(dataframe["VMAF_score"])  
    MSE=mean_squared_error(x, y)
    print("RMSE:",math.sqrt(MSE))
    print("R2:",1-MSE/np.var(np.array(x)))
    plt.scatter(x, y, marker='o', label=content[a])
    
    if b is not None:
        dataframe=sample_data.loc[sample_data["content_id"] == b,["DMOS","VMAF_score","VMAF_feature_motion2_score"]] 
        print(dataframe[1:].describe())
        x=list(dataframe["DMOS"])
        y=list(dataframe["VMAF_score"])
        MSE=mean_squared_error(x, y)
        print("RMSE:",math.sqrt(MSE))
        print("R2:",1-MSE/np.var(np.array(x)))
        plt.scatter(x, y, marker='o', label=content[b])
    if c is not None:
        dataframe=sample_data.loc[sample_data["content_id"] == c,["DMOS","VMAF_score","VMAF_feature_motion2_score"]] 
        print(dataframe[1:].describe())
        x=list(dataframe["DMOS"])
        y=list(dataframe["VMAF_score"])
        MSE=mean_squared_error(x, y)
        print("RMSE:",math.sqrt(MSE))
        print("R2:",1-MSE/np.var(np.array(x)))
        plt.scatter(x, y, marker='o', label=content[c])

    plt.legend(loc = "best" )#图例
    plt.grid()
    plt.plot([0,105], [0,105])
    plt.show()
    
```

```python

drawScatter(sample_data,0,5)

#结论：CG vs Animation

#VMAF评分范围一致，在25-98之间.DMOS中，Animation的人类感觉起点更高。在不清晰的情况下，人的感受对Animation包容度更高。
#在清晰的情况（70以上），CG动画会VMAF打分都偏低的情况更多
#猜想：motion导致偏离？

```

```python
drawScatter(sample_data,1,8)
#对motion开始研究？ 

#结论： 不动的有object的物体 VS 动的有object物体
#对object的评分有明显的分水岭，70之前偏高，70之后偏低。且不动的物体在人体感官基本一致的情况下，打分VMAF的分数明显较高，应该是考虑了motion的原因
#会动的高显著性object，高编码时评分基本正确（考虑了motion），但低编码时偏差较大，motion的加入有优势有劣势（唯一不同，有motion）
#原因猜想？motion会另运动的物体评分总体偏高。
```

```python
drawScatter(sample_data,8,5)


#前面的猜想：motion会另运动的物体评分总体偏高似乎是错误的。研究两个motion一致的物体，可见对于不同类型的视频，motion会另高编码的视频的评分更准确，但在
#结论：动的动画VS 动的人
#低编码时却有不同的影响，两极分化非常严重。
#原因猜想？动画与视频的亮度，色彩丰富度原因。由于动画色彩丰富，DMOS比模型评分更高。

```

```python
drawScatter(sample_data,2,7,3)

#结论： 动的无object VS 动的无object
#范围差不多，对于人群的评分准确率比较高，
#对比较近的人群评分低感觉高，对人群比较远的评分高、感觉低，总体问题不大，因为motion非常高，对于这种没有object的视频，motion非常有用。
```

```python
drawScatter(sample_data,2,3)


#结论：动的有主体 VS 动的无主体
#都是拥有人群的画面，motion的分数也基本一致，但是人群中有一个主体时，分数更加准确。（黄色基本都压线）
```

```python
drawScatter(sample_data,6,2)
#结论： 不动的无object VS  动的无object的
#运动的无焦点的人群，准确度较高。不运动的无焦点的尝试，评分较感觉更低，
#缺乏motion的计算，加上色彩较暗
```

```python
drawScatter(sample_data,4)
#结论：拍摄手法一致，但是有water的视频 打分比感觉低，噪音过多造成的评分差距较大，人的视觉感觉会受影响，人会适当忽略。
```

```python
drawScatter(sample_data,1,6)

#结论:不动的无objectVS不动的有object
#有object的时候，低编码时比较准确，高编码时普遍VMAF偏低
```

# 总结：

> motion 无object crowd/crowd-closer 准确
> motion 无object VS motion 有object crowd/show 基本准确，有object会更准确


> motion 有object animation/Tennis 不准确 低编码时 偏差较大 (animation除外，有色彩影响，另一个并到下方)
> motion 有object VS 无motion 有object tennis/bird 低编码时评分偏高


> 无motion 无object VS 无motion 有object  Old-city/bird 不准确 高编码时，高普遍VMAF偏低
> motion 无object VS 无motion 无object cownd/old-city 后者不准确，高编码时评分较低（只有后者，所以可以并到上一个case）


### 高噪音和动画需要提高VMAF分数,动画色彩和亮度丰富；高噪音则是人类有包容性
### 准确的情况：光有Motion无object时
### 不准确的情况：低编码评分偏高，有object，有无motion
### 不准确的情况：高编码评分偏低，无motion，有无object
